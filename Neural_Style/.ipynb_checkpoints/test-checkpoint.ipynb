### import pytorch

```python
import torch
torch.cuda.is_available()
```

### import model and read the weights

```python
from transformer_net import TransformerNet
```

```python
model = TransformerNet()
model.load_state_dict(torch.load('checkpoints/GodBearer.pth'))
model.cuda();
model.eval();
```

### convert image from opencv to PIL

```python
from PIL import Image
from torchvision import transforms
import cv2
```

```python
img = cv2.cvtColor(cv2.imread('outputs/test.jpg'), cv2.COLOR_BGR2RGB)
img = Image.fromarray(img).resize((480, 270))
img
```

### define torchvision transform

```python
trm = transforms.Compose([
    transforms.ToTensor(),
    transforms.Lambda(lambda x: x*255)
])
```

### forward pass and transform to PIL Image

```python
img = trm(img).cuda()
t_img = model(img.unsqueeze(0)).squeeze(0).cpu()

t_img /= 255
t_img[t_img > 1] = 1
t_img[t_img < 0] = 0
img = transforms.ToPILImage()(t_img)
img
```

### convert PIL image to numpy array (opencv image)

```python
import numpy as np
img = np.array(img)
```
